```python
import sys
import pandas as pd
import sklearn
from sklearn.preprocessing import StandardScaler
import numpy as np
from numpy import sort
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn import svm
```

```python
dataset_path = 'C:/Users/nukhba/master-thesis/data/csv/diff_dataset.csv'
df = pd.read_csv(dataset_path)
df.head()
```

```python
df.shape
```

```python
column = df.columns
#df.drop(columns=['Unnamed: 0'], inplace=True)

# Separating out the features
x = df[column[:-1]]
```

```python
# Standardizing the features
x_normalized = StandardScaler().fit_transform(x)
scaled_features_df = pd.DataFrame(x_normalized, index=x.index, columns=x.columns)
scaled_features_df
```

```python
from sklearn.cluster import KMeans
 
#Initialize the class object,, setting k = 3
kmeans = KMeans(n_clusters= 3)

clusters = kmeans.fit_predict(scaled_features_df)
print(clusters)
```
