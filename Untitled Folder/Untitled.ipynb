```python
from sklearn.preprocessing import StandardScaler

column = dataset.columns
# Separating out the features
x = dataset[column[:-1]]

# Separating out the target
y = dataset[column[-1]]

print(x.shape, y.shape)
```

```python
# Standardizing the features
x_normalized = StandardScaler().fit_transform(x.values)
scaled_features_df = pd.DataFrame(x_normalized, index=x.index, columns=x.columns)
scaled_features_df
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plot

pca = PCA(n_components=x.shape[1])
pca.fit(scaled_features_df)

# Reformat and view results
loadings = pd.DataFrame(pca.components_.T,
columns=['PC%s' % _ for _ in range(len(scaled_features_df.columns))],
index=x.columns)
print(loadings)

plot.plot(pca.explained_variance_ratio_)
plot.ylabel('Explained Variance')
plot.xlabel('Components')
plot.show()
```

```python
loadings = pca.components_
num_pc = pca.n_features_
np.cumsum(pca.explained_variance_ratio_)
pc_list = ["PC"+str(i) for i in list(range(1, num_pc+1))]
loadings_df = pd.DataFrame.from_dict(dict(zip(pc_list, loadings)))
loadings_df['variable'] = scaled_features_df.columns.values
loadings_df = loadings_df.set_index('variable')
loadings_df
```
